# Concatenate

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({"Name": ["Vinayak", "Amit", "Sarva", "Ravi", "Vassu"], "Suranme": ["Ushakola", "Jha", "Palaniswami", "Sonkar", "Vasala"]})
df1
```

```python
df2 = pd.DataFrame({"Name": ["Sidhu", "Rohit"], "Suranme": ["Ambati", "Bagdi"]})
df2
```

```python
new_df = pd.concat([df1, df2], axis=0)
new_df
```

```python
# To cancat rows
new_df = pd.concat([df1, df2], axis=0, ignore_index=True)
new_df
```

```python
df3 = pd.DataFrame({"Age": [22, 22 ,23, 21, 23, 21, 24]})
df3
```

```python
# To concat columns
new_df = pd.concat([new_df, df3], axis=1)
new_df
```
